```python
# 과대적합과 과소적합 발생
# 두 문제가 일어났을 때 해결하는 방법들 사용
# 해결 방법으로 과대적합에는 규제, 드롭아웃을 사용
```

```python
from __future__ import absolute_import, division, print_function, unicode_literals, unicode_literals

import numpy as np
import matplotlib.pyplot as plt
```

```python
import tensorflow as tf
from tensorflow import keras

print(tf.__version__)
```

```python
# IMDB 데이터셋 다운로드

NUM_WORDS = 10000

(train_data, train_labels), (test_data, test_labels) = keras.datasets.imdb.load_data(num_words=NUM_WORDS)
```

```python
# 이전처럼 문장을 임베딩하지 않음
# 문장을 멀티 핫 인코딩으로 변환 -> 단어가 나타났으면 1 아니면 0
# 빠르게 과대적합된다.
def multi_hot_sequence(sequences, dimension):
    # 0으로 채워진 (sequence 길이, dimension)인 행렬로 만듬
    results = np.zeros((len(sequences), dimension))

    for i, word_indices in enumerate(sequences):
        results[i, word_indices] = 1.0
    return results
```

```python
train_data = multi_hot_sequence(train_data, dimension=NUM_WORDS)
test_data = multi_hot_sequence(test_data, dimension=NUM_WORDS)
```

```python
# 나타난 단어들을 시각적으로 확인
plt.plot(train_data[0])
```

```python
# 과대적합 예제
# 가장 쉬운 과대적합 예방 -> 모델의 규모 축소

# 기준 모델
baseline_model = keras.Sequential([
    # '.summary' 메서드로 인해서 'input_shape' 필요
    keras.layers.Dense(16, activation='relu', input_shape=(NUM_WORDS,)),
    keras.layers.Dense(16, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

baseline_model.compile(optimizer='adam',
                      loss='binary_crossentropy',
                      metrics=['accuracy', 'binary_crossentropy'])

baseline_model.summary()
```

```python
# 기준 모델 학습
baseline_model_history = baseline_model.fit(train_data,
                                           train_labels,
                                           epochs=20,
                                           batch_size=512,
                                           validation_data=(test_data, test_labels),
                                           verbose=2)
```

```python
# 작은 모델
smaller_model = keras.Sequential([
    keras.layers.Dense(4, activation='relu', input_shape=(NUM_WORDS,)),
    keras.layers.Dense(4, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

smaller_model.compile(optimizer='adam',
                    loss='binary_crossentropy',
                    metrics=['accuracy', 'binary_crossentropy'])

smaller_model.summary()
```

```python
# 작은 모델 훈련
smaller_model_history = smaller_model.fit(train_data,
                                         train_labels,
                                         epochs=20,
                                         batch_size=512,
                                         validation_data=(test_data, test_labels),
                                         verbose=2)
```

```python
# 큰 모델
bigger_model = keras.models.Sequential([
    keras.layers.Dense(512, activation='relu', input_shape=(NUM_WORDS,)),
    keras.layers.Dense(512, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

bigger_model.compile(optimizer='adam',
                     loss='binary_crossentropy',
                     metrics=['accuracy','binary_crossentropy'])

bigger_model.summary()
```

```python
# 큰 모델 훈련
bigger_model_history = bigger_model.fit(train_data, train_labels,
                                  epochs=20,
                                  batch_size=512,
                                  validation_data=(test_data, test_labels),
                                  verbose=2)
```

```python
# 훈련 손실 및 검증 손실 그래프
# 훈련값은 계속해서 내려가지만 validation 값은 계속해서 올라가는 과대적합을 보여준다.

def plot_history(histories, key = 'binary_crossentropy'):
    plt.figure(figsize=(16,10))
    
    # 각 히스토리들에서 검증 값은 점선으로 표시
    # 각 히스토리들의 훈련 값은 실선으로 표시
    for name, history in histories:
        val = plt.plot(history.epoch, history.history['val_'+key],
                       '--', label=name.title()+' Val')
        plt.plot(history.epoch, history.history[key], color=val[0].get_color(),
                 label=name.title()+' Train')
        
    plt.xlabel('Epochs')
    plt.ylabel(key.replace('_', ' ').title())
    plt.legend()
    
    plt.xlim([0, max(history.epoch)])
    
plot_history([('baseline', baseline_model_history),
              ('smaller', smaller_model_history),
              ('bigger', bigger_model_history)])
```

```python
# 과대 적합을 방지하기 위해서 가중치를 규제하기
# L2 규제를 추가

l2_model = keras.models.Sequential([
    keras.layers.Dense(16, kernel_regularizer=keras.regularizers.l2(0.001),
                       activation='relu', input_shape=(NUM_WORDS,)),
    keras.layers.Dense(16, kernel_regularizer=keras.regularizers.l2(0.001),
                       activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

l2_model.compile(optimizer='adam',
                 loss='binary_crossentropy',
                 metrics=['accuracy', 'binary_crossentropy'])
```

```python
# L2 모델 훈련
l2_model_history = l2_model.fit(train_data, train_labels,
                                epochs=20,
                                batch_size=512,
                                validation_data=(test_data, test_labels),
                                verbose=2)
```

```python
# L2 규제의 성능 시각화
plot_history([('baseline', baseline_model_history),
              ('l2', l2_model_history)])
# 과적합이 많이 줄었다.
```

```python
# 드롭아웃 추가
dpt_model = keras.models.Sequential([
    keras.layers.Dense(16, activation='relu', input_shape=(NUM_WORDS,)),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(16, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1, activation='sigmoid')
])

dpt_model.compile(optimizer='adam',
                  loss='binary_crossentropy',
                  metrics=['accuracy','binary_crossentropy'])
```

```python
# 드롭아웃 훈련
dpt_model_history = dpt_model.fit(train_data, train_labels,
                                  epochs=20,
                                  batch_size=512,
                                  validation_data=(test_data, test_labels),
                                  verbose=2)
```

```python
# 드롭아웃 모델 성능 시각화
plot_history([('baseline', baseline_model_history),
              ('dropout', dpt_model_history)])
```
