# 1. 합성곱 신경망 소개

- Dense로 풀었던 MNIST 숫자 이미지 분류는 약 97.8%의 정확도를 가졌다.
- 이번에는 MNIST 숫자 이미지 분류에 Conv넷을 적용 해볼것이다.


### 간단한 컨브넷 만들기

```python
from keras import layers
from keras import models

model = models.Sequential()
model.add(layers.Conv2D(32, (3,3), activation='relu', input_shape=(28, 28, 1)))
model.add(layers.MaxPool2D((2,2)))
model.add(layers.Conv2D(64, (3,3), activation='relu'))
model.add(layers.MaxPool2D((2,2)))
model.add(layers.Conv2D(64, (3,3), activation='relu'))
```

- 컨브넷은 (image_height, image_width, image_channels) 크기의 입력 텐서를 사용한다.


### 위 컨브넷 구조 출력

```python
model.summary()
```

- Conv2D와 MaxPooling2D 층의 출력은 (height, width, channels) 크기의 3D 텐서이다.
- 다음 단계에는 출력 텐서를 Dense넷에 주입한다.
- Dense넷에 입력하기 위해서 1차원 벡터로 만들어준다.


### 컨브넷 위에 분류기 추가하기

```python
model.add(layers.Flatten()) # 1차원 벡터로 변환한다.
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(10, activation='softmax')) # 어떤 숫자인지 판별해야하므로 softmax를 사용하여 확률을 구함
```

### 전체 네트워크 보기

```python
model.summary()
```

- (3,3,64)인 출력물이 (576)인 벡터로 펼쳐져 Dense 넷에 입력되는것을 알 수 있다.


### MNIST 이미지에 컨브넷 훈련하기

```python
from keras.datasets import mnist
from keras.utils import to_categorical

(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

# train_images는 채널을 가지고 있지 않으므로 채널을 만들어 주고, 값을 정규화 시켜준다.
train_images = train_images.reshape((60000, 28, 28, 1))
train_images = train_images.astype('float32') / 255

test_images = test_images.reshape((10000, 28, 28, 1))
test_images = test_images.astype('float32') / 255

# 라벨들을 원핫 인코딩으로 변경
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)

model.compile(optimizer='rmsprop',
             loss='categorical_crossentropy',
             metrics=['accuracy'])

model.fit(train_images, train_labels, epochs=5, batch_size=64)
```

### 테스트 데이터에서 모델을 평가

```python
test_loss, test_accuracy = model.evaluate(test_images, test_labels)
print("테스트 데이터 정확도: {}".format(test_accuracy))
```

- 간단한 모델을 만들었음에도 모델의 정확도가 99.3% 정도가 나왔다.
- 다음에는 왜 컨브넷이 더 잘 작동하는지 알아본다.

## 1.1. 합성곱 연산

- Dense 층 -> 입력 특성 공간에 있는 전역 패턴을 학습
- Conv 층 -> 2D 윈도우로 입력에서 패턴을 찾는다.


- 위 핵심 특징이 컨브넷에 제공하는 성질]
    1. 학습된 패턴은 평행 이동 불변성을 가진다.( 어떤 패턴을 학습했다면 다른 곳에서도 패턴을 인식할 수 있다.)
    2. 패턴의 공간적 계층 구조를 학습할 수 있다.(첫 번째 층은 에지, 두 번째 층은 더 큰 패턴을 학습하는 식으로 복잡하고 추상적인 시각적 개념을 효과적으로 학습할 수 있다.)
    
    
- 합성곱의 핵심 파라미터
    1. 입력으로부터 뽑아낼 패치 크기 : 전형적으로 3X3 5X5 크기글 사용
    2. 특성 맵의 출력 깊이 : 합성곱으로 계산할 필터의 수
    
    
- 합성곱의 작동 방식 예시

### 경계 문제와 패딩 이해

- conv 넷을 적용시 출력의 크기는 입력 데이터의 크기보다 줄어들게 되어있다.
- 출력 크기를 변화시키고 싶지 않다면 패딩을 사용할 수 있다.
- keras에서 패딩 인자로 valid는 패딩을 사용하지 않는다. same은 입력과 동일한 높이를 가진 출력을 만드는 패딩을 사용한다가 된다.

## 1.2. 최대 풀링 연산

1. 다운 샘플링 역활을 맡는다.
2. 특성 맵의 가중치 개수를 줄인다.

## 2.3. 네트워크 구성하기

- MNIST 숫자 이미지보다 이미지가 크고 복잡한 문제이기 때문에 네트워크를 좀 더 크게 만든다.
- 이진 분류 문제이므로 네트워크는 하나의 유닛과 sigmoid 활성화 함수로 끝난다.
- 이미지는 150 * 150인 RGB 채널을 가진 이미지로 변형 시킬것이다..

```python
model = models.Sequential()
model.add(layers.Conv2D(32, (3,3), activation='relu', input_shape=(150, 150, 3)))
model.add(layers.MaxPool2D((2,2)))
model.add(layers.Conv2D(64, (3,3), activation='relu'))
model.add(layers.MaxPool2D((2,2)))
model.add(layers.Conv2D(128, (3,3), activation='relu'))
model.add(layers.MaxPool2D((2,2)))
model.add(layers.Conv2D(128, (3,3), activation='relu'))
model.add(layers.MaxPool2D((2,2)))
model.add(layers.Flatten())
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

### 컴파일 단계 

- RMSprop 옵티마이저 선택
- 이진 크로스엔트로피 손실 사용

```python
from keras import optimizers

model.compile(optimizer=optimizers.rmsprop(lr=1e-4),
             loss='binary_crossentropy',
             metrics=['acc'])
```

## 2.4. 데이터 전처리

- 데이터는 네트워크에 주입되기 전에 부동 소수 타입 텐서로 전처리 되어야한다.


- JPEG 파일 전처리 과정
    1. 사진 파일 읽기
    2. JPEG 콘텐츠를 RGB 픽셀 값으로 디코딩
    3. 부동 소수 타입의 텐서로 변환
    4. 픽셀 값(0 ~ 255)를 [0~1] 사이로 조정
    

- 위와 같은 전처리를 도와주기 위해서 케라스에는 keras.preprocessing.image에서 헬퍼 도구를 제공한다.
- 특히 ImageDataGenerator는 이미지파일을 배치 텐서로 자동으로 만들어주며 레이블도 제공해준다.
- ImageDataGenerator는 pillow 라이브러리를 사용하므로 없다면 설치해놓아야 한다.

```python
from keras.preprocessing.image import ImageDataGenerator

# 이미지 스케일 조정
train_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(train_dir,
                                                   target_size = (150, 150),
                                                   batch_size=20,
                                                   class_mode='binary')

validation_generator = test_datagen.flow_from_directory(validation_dir,
                                                 target_size=(150, 150),
                                                       batch_size=20,
                                                       class_mode='binary')
```

### 배치 제네레이터를 사용하여 모델 훈련

```python
history = model.fit_generator(train_generator,
                             steps_per_epoch=100,
                             epochs=30,
                             validation_data=validation_generator,
                             validation_steps=50)
```

```python
# 학습이 끝난 모델을 저장
model.save('cats_and_dogs_small_1.h5')
```

- 훈련, 검증 데이터에 대한 모델의 손실과 정확도 그래프 생성

```python
import matplotlib.pyplot as plt

acc = history.history['acc']
val_acc = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(1, len(acc) + 1)

plt.plot(epochs, acc, 'bo', label='Training acc')
plt.plot(epochs, val_acc, 'b', label='Validation acc')
plt.title('Training and validation accuracy')
plt.legend()

plt.figure()

plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.legend()

plt.show()
```

- 위 그래프를 보면 모델이 과대적합하는것을 볼 수 있다.

## 2.5. 데이터 증식 사용하기

- ImageDataGenerator에서 읽은 이미지에 여러 종류의 랜덤 변환을 적용할 수 있으므로 그것을 사용한다.

```python
datagen = ImageDataGenerator(rotation_range=20,
                            width_shift_range=0.1,
                            height_shift_range=0.1,
                            shear_range=0.1,
                            zoom_range=0.1,
                            horizontal_flip=True,
                            fill_mode='nearest')
```

- 위 코드에서 나온 매개변수 설명
    1. rotation_range -> 랜덤하게 사진을 회전시킬 각도 범위(0 ~ 180 사이)
    2. *_shift_range -> 사진을 수평과 수직으로 랜덤하게 평행 이동 시킬 범위
    3. shear_range -> 랜덤하게 전단 변환을 적용할 각도 범위
    4. zoom_range -> 랜덤하게 사진을 확대할 범위
    5. horizontal_flip -> 랜덤하게 이미지를 수평으로 뒤집는다.
    6. fill_mode -> 회전이나 가로/세로 이동으로 채워야하는 픽셀을 채울 전략
    
    
### 랜덤하게 증식된 훈련 이미지 그리기

```python
from keras.preprocessing import image

# 폴더에 있는 고양이 사진의 경로를 모두 리스트화해서 저장
fnames = sorted([os.path.join(train_cats_dir, fname) for fname in os.listdir(train_cats_dir)]) 

img_path = fnames[3] # 증식할 이미지 선택

img = image.load_img(img_path, target_size=(150, 150)) # 이미지를 읽고 크기를 변형

x = image.img_to_array(img) # 이미지를 넘파이 배열로 변경
x = x.reshape((1,) + x.shape) # 크기를 (1, 150, 150, 3)으로 변경함

i = 0
# 랜덤하게 변환된 이미지 배치 생성
for batch in datagen.flow(x, batch_size=1):
    plt.figure(i)
    imgplot = plt.imshow(image.array_to_img(batch[0]))
    i += 1
    if i % 4 == 0:
        break
        
plt.show()
```

- 데이터 증식을 사용하여 새로운 네트워크를 훈련시키면 네트워크에 같은 입력 데이터가 두 번 주입되지 않는다.
- 하지만 데이터간의 상호 연관성이 높다.
- 과대적합을 억제하기 위해서 Dropout 층을 추가

```python
model = models.Sequential()
model.add(layers.Conv2D(32, (3,3), activation='relu', input_shape=(150, 150, 3)))
model.add(layers.MaxPool2D((2,2)))
model.add(layers.Conv2D(64, (3,3), activation='relu'))
model.add(layers.MaxPool2D((2,2)))
model.add(layers.Conv2D(128, (3,3), activation='relu'))
model.add(layers.MaxPool2D((2,2)))
model.add(layers.Conv2D(128, (3,3), activation='relu'))
model.add(layers.MaxPool2D((2,2)))
model.add(layers.Flatten())
model.add(layers.Dropout(0.5))
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy',
             optimizer=optimizers.RMSprop(lr=1e-4),
             metrics=['acc'])
```

### 데이터 증식과 드롭아웃을 사용하여 네트워크 훈련

```python
train_datagen = ImageDataGenerator(rescale=1./255,
                            rotation_range=40,
                            width_shift_range=0.2,
                            height_shift_range=0.2,
                            shear_range=0.2,
                            zoom_range=0.2,
                            horizontal_flip=True)
```

```python
test_datagen = ImageDataGenerator(rescale=1./255) # 검증 데이터는 증식되면 안된다.

train_generator = train_datagen.flow_from_directory(train_dir,
                                                   target_size=(150, 150),
                                                   batch_size=32,
                                                   class_mode='binary')

validation_generator = test_datagen.flow_from_directory(validation_dir,
                                                       target_size=(150, 150),
                                                       batch_size=32,
                                                       class_mode='binary')

history = model.fit_generator(train_generator,
                             steps_per_epoch=100,
                             epochs=100,
                             validation_data=validation_generator,
                             validation_steps=50)
```

```python
# 학습된 모델 저장

model.save('cats_and_dogs_small_2.h5')
```
