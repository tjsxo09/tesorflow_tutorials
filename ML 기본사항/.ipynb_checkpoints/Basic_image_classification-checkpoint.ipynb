```python
from __future__ import absolute_import, division, print_function, unicode_literals, unicode_literals

# tensorflow, tf.keras import
import tensorflow as tf
from tensorflow import keras
```

```python
# 헬퍼 라이브러리 임포트
import numpy as np
import matplotlib.pyplot as plt

print(tf.__version__) # 파이썬 버전 출력
```

```python
# 옷과 신발 이미지가 들어있는 패션 mnist 데이터를 받아옴
fashion_mnist = keras.datasets.fashion_mnist

(train_image, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

```python
# label에 1:1 매칭되는 클래스 이름들을 설정
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
print(train_image.shape) # 60000개의 28*28*1 이미지 데이터가 들어있음
```

```python
print(train_labels.shape) # training 데이터 60000개에 대해서 라벨링이 되어있음
```

```python
plt.figure()
plt.imshow(train_image[0]) # 0번째 이미지를 출력함
plt.colorbar() # 값의 범위를 보여주는 막대를 표시
plt.grid(False) # 격자 그리드는 출력하지 않음
plt.show()
```

```python
# 0~255 사이인 값이기 때문에 0~1사이의 값으로 정규화를 해준다.

train_image = train_image / 255.0
test_images = test_images / 255.0
```

```python
plt.figure(figsize=(10,10))
# 25개의 훈련 이미지 출력
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([]) # x축 범위를 표시안하게 만듬
    plt.yticks([]) # y축 범위를 표시안하게 만듬
    plt.grid(False)
    plt.imshow(train_image[i], cmap=plt.cm.binary)
    plt.xlabel(class_names[train_labels[i]])
plt.show()
```

```python
# 층 설정
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])
```

```python
# 모델 컴파일
# 1. 손실함수 설정
# 2. 옵티마이저 설정
# 3. Metrics 설정 -> 테스트 단계 모니터링에 사용 됨

model.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])
```

```python
# 모델 훈련
# 모델 훈련에 필요한 입력 데이터, 입력 데이터의 라벨, 전체 데이터 훈련 횟수를 넣어준다.
model.fit(train_image, train_labels, epochs=5)
```

```python
# 정확도 평가
# 모델 평가에 필요한 입력 데이터, 입력 데이터의 라벨, 평가 표시 옵션을 넣어준다.
test_loss, test_acc = model.evaluate(test_images, test_labels, verbose=2)

print('\n정확도:', test_acc)
```

```python
# 예측값 만들기
# 전체 이미지에 대한 예측값을 반환
predictions = model.predict(test_images)
```

```python
print(predictions[0]) # 10개 라벨에 대한 값들이 출력된다.
```

```python
np.argmax(predictions[0]) # 10개 라벨중에서 가장 값이 높은것을 반환 -> 가장 확률이 높은것
```

```python
 test_labels[0] # 에측값은 9 테스트 라벨도 9 이므로 잘 예측했다고 표현할 수 있다.
```

```python
# 10개 클래스에 대한 예측을 그래프로 표현
# plt에 이미지를 등록하는 함수
def plot_image(i, predictions_array, true_label, img):
    predictions_array, true_label, img = predictions_array[i], true_label[i], img[i]
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    
    # img를 출력
    plt.imshow(img, cmap=plt.cm.binary)
    
    predicted_label = np.argmax(predictions_array)
    
    if predicted_label == true_label:
        color = 'blue'
    else:
        color = 'red'
    
    # xlabel에서 출력할 값 예측한 이름, 예측 확률, 진자 이름
    # 맞으면 파란색 틀리면 빨간색으로 출력
    plt.xlabel("{} {:2.0f}% ({})".format(class_names[predicted_label],
                                        100*np.max(predictions_array),
                                        class_names[true_label]),
                color=color)
# plt
def plot_value_array(i, predictions_array, true_label):
    predictions_array, true_label = predictions_array[i], true_label[i]
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    
    # 예측값들의 확률 크기를 보여줄 막대 그래프
    thisplot = plt.bar(range(10), predictions_array, color="#777777")
    plt.ylim([0,1]) # y값은 0~1 사이이다.
    
    # 예측 값
    predicted_label = np.argmax(predictions_array)
    
    # 예측값은 붉은색으로 실제값은 파랑으로 설정
    # 예측값과 실제값이 같으면 파랑 하나만이 남는다.
    thisplot[predicted_label].set_color('red')
    thisplot[true_label].set_color('blue')
```

```python
i = 0
plt.figure(figsize=(6,3))
plt.subplot(1,2,1)
plot_image(i, predictions, test_labels, test_images)
plt.subplot(1,2,2)
plot_value_array(i, predictions, test_labels)
plt.show()
```

```python
i = 12
plt.figure(figsize=(6,3))
plt.subplot(1,2,1)
plot_image(i, predictions, test_labels, test_images)
plt.subplot(1,2,2)
plot_value_array(i, predictions, test_labels)
plt.show()
```

```python
num_rows = 5
num_cols = 3

# 이미지의 개수는 세로와 가로를 곱한것
num_images = num_rows*num_cols
plt.figure(figsize=(2*2*num_cols, 2*num_rows))

# 총 5*3의 이미지 개수 만큼 반복
for i in range(num_images):
    # 물건 이미지와 확률 값을 보여주는 막대 그래프가 필요하기 때문에
    # 세로는 row 만큼 분할 가로는 cols*2개 만큼 분할
    # 마지막 2*i+1은 생성된 subplot 순서를 의미
    plt.subplot(num_rows, 2*num_cols, 2*i+1)
    plot_image(i, predictions, test_labels, test_images)
    plt.subplot(num_rows, 2*num_cols, 2*i+2)
    plot_value_array(i, predictions, test_labels)
plt.show()
```
