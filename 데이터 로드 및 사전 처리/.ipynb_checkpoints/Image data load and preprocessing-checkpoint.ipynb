```python
from __future__ import absolute_import, division, print_function, unicode_literals
```

```python
import tensorflow as tf
```

```python
AUTOTUNE = tf.data.experimental.AUTOTUNE
```

```python
import IPython.display as display
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
tf.__version__
```

```python
import pathlib
# get flower photos data from Google URL
# file is named flower_photos
# untar은 압축 해제 한 상태로 데이터를 가져오는것을 의미한다.
data_dir = tf.keras.utils.get_file(origin='https://storage.googleapis.com/download.tensorflow.org/example_images/flower_photos.tgz',
                                         fname='flower_photos', untar=True)
# pathlib는 주소를 하나의 객체로 본다.
# tf.keras.utils.get_file은 얻은 파일의 주소를 반환한다.
data_dir = pathlib.Path(data_dir)
```

```python
# count number of images
# .jpg로 끝나는 모든 파일이름을 저장한 후 길이를 재는것으로 이미지의 개수를 파악한다.
image_count = len(list(data_dir.glob('*/*.jpg')))
image_count
```

```python
# LICENSE.txt와 다른 모든 파일, 디렉토리 이름을 CLASS_NAMES로 저장한다.
CLASS_NAMES = np.array([item.name for item in data_dir.glob('*') if item.name != "LICENSE.txt"])
print(CLASS_NAMES)
```

```python
roses = list(data_dir.glob('roses/*'))

for image_path in roses[:3]:
    display.display(Image.open(str(image_path)))
```

```python
# 1.Load using keras.preprocessing

# The 1./255 is to convert from uint8 to float32 in range [0,1]
# 학습을 좀 더 정확하고 빠르게 하기 위해서 rescale을 사용한다.
image_generator = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1./255)
```

```python
BATCH_SIZE = 32
IMG_HEIGHT = 224
IMG_WIDTH = 224
# np.ceil은 크거나 같은 값으로 반환 함
STEPS_PER_EPOCH = np.ceil(image_count/BATCH_SIZE)
```

```python
# generate train data
# from directory generate train data
train_data_gen = image_generator.flow_from_directory(directory=str(data_dir),
                                                    batch_size=BATCH_SIZE,
                                                    shuffle=True,
                                                    target_size=(IMG_HEIGHT, IMG_WIDTH),
                                                    classes=list(CLASS_NAMES))
```

```python
def show_batch(image_batch, label_batch):
    plt.figure(figsize=(10, 10))
    for n in range(25):
        ax = plt.subplot(5, 5, n+1)
        plt.imshow(image_batch[n])
        plt.title(CLASS_NAMES[label_batch[n]==1][0].title())
        plt.axis('off')
```

```python
image_batch, label_batch = next(train_data_gen)
show_batch(image_batch, label_batch)
```

```python
# 2.Load using tf.data
# tf.data를 이용한 방식에는 몇가지 문제가 있다.
# 1. 느리다.
# 2. 세밀한 컨트롤에 문제가 있다.
# 3. Tensorflow와 잘 통합되지 않는다.

# tf.data를 이용한 전처리 방식은 tf.keras.preprocessing과 비교를 위해 사용되므로 코드를 복붙했습니다.
```

```python
list_ds = tf.data.Dataset.list_files(str(data_dir/'*/*'))
```

```python
for f in list_ds.take(5):
    print(f.numpy())
```

```python
def get_label(file_path):
    # convert the path to a list of path components
    parts = tf.strings.split(file_path, os.path.sep)
    # The second to last is the class-directory
    return parts[-2] == CLASS_NAMES
```

```python
def decode_img(img):
    # convert the compressed string to a 3D uint8 tensor
    img = tf.image.decode_jpeg(img, channels=3)
    # Use `convert_image_dtype` to convert to floats in the [0,1] range.
    img = tf.image.convert_image_dtype(img, tf.float32)
    # resize the image to the desired size.
    return tf.image.resize(img, [IMG_WIDTH, IMG_HEIGHT])
```

```python
def process_path(file_path):
    label = get_label(file_path)
    # load the raw data from the file as a string
    img = tf.io.read_file(file_path)
    img = decode_img(img)
    return img, label
```

```python
# Set `num_parallel_calls` so multiple images are loaded/processed in parallel.
labeled_ds = list_ds.map(process_path, num_parallel_calls=AUTOTUNE)
```

```python
for image, label in labeled_ds.take(1):
    print("Image shape: ", image.numpy().shape)
    print("Label: ", label.numpy())
```

```python
# preparing data for tf.data

def prepare_for_training(ds, cache=True, shuffle_buffer_size=1000):
    #.cache(filename)을 이용해서 캐시 전처리 작업을 수행한다.
    # 메모리에 최적화 되어있지 않다.
    
    if cache:
        if isinstance(cache, str):
            ds = ds.cache(cache)
        else:
            ds = ds.cache()
            
    ds = ds.shuffle(buffer_size=shuffle_buffer_size)
    
    # 계속 반복
    ds = ds.repeat()
    
    ds = ds.batch(BATCH_SIZE)
    
    # prefetch는 모델이 학습중일때 백그라운드에서 데이터셋을 배치에 맞게 맞춰준다.
    ds = ds.prefetch(buffer_size=AUTOTUNE)

    return ds            
```

```python
train_ds = prepare_for_training(labeled_ds)

image_batch, label_batch = next(iter(train_ds))
```

```python
show_batch(image_batch.numpy(), label_batch.numpy())
```

```python
# Performance 측정
import time
# 1000ms is 1s
default_timeit_steps = 1000
```

```python
def timeit(ds, steps=default_timeit_steps):
    start = time.time()
    it = iter(ds)
    for i in range(steps):
        batch = next(it)
        if i%10 == 0:
            print('.', end='')
    print()
    end = time.time()
        
    duration = end-start
    print("{} batches: {} s".format(steps, duration))
    print("{:0.5f} Images/s".format(BATCH_SIZE*steps/duration))
```

```python
# keras 전처리에 걸리는 시간
timeit(train_data_gen)
```

```python
# tf.data 전처리에 걸리는 시간
# cache를 사용하여 tf.keras보다도 빠르게 전처리가 가능하다.
timeit(train_ds)
```

```python
# cache를 사용하지 않았을때 걸리는 시간
uncached_ds = prepare_for_training(labeled_ds, cache=False)
timeit(uncached_ds)
```

```python
# 데이터 셋이 메모리에 들어가지 않으면 cache를 사용하여 전처리 할 수 있다.
filecache_ds = prepare_for_training(labeled_ds, cache="./flowers.tfcache")
timeit(filecache_ds)
```
