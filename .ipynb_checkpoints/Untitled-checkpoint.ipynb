```python
from __future__ import absolute_import, division, print_function, unicode_literals, unicode_literals

import pathlib
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

print(tf.__version__)
```

```python
# 데이터 셋 다운로드
# Auto-mpg dataset
# 1970~1980년대 초반의 자동차 연비 데이터
dataset_path = keras.utils.get_file("auto-mpg.data", "http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data")
dataset_path
```

```python
# 데이터의 속성 이름들
column_names = ['MPG', 'Cylinders', 'Displacement', 'Horsepower', 'Weight',
                'Acceleration', 'Model Year', 'Origin']

# pandas를 이용하여 데이터를 읽음
raw_datatset = pd.read_csv(dataset_path, names=column_names,
                           na_values = "?", comment='\t',
                           sep=" ", skipinitialspace=True)

dataset = raw_datatset.copy()
dataset.tail()
```

```python
# 데이터셋에 누락된 부분이 있는가 확인
# Horsepower(마력) 부분에 누락이 총 6개 있음
dataset.isna().sum()
```

```python
# 누락된 행 삭제
dataset = dataset.dropna()
dataset.isna().sum()
```

```python
# origin 속성은 수치값이 아닌 범주형이므로 원-핫 인코딩으로 변환
# origin은 1, 2, 3 3개의 범주값이 있다
# 1 = USA, 2 = Europe, 3 = Japan
origin = dataset.pop('Origin')
```

```python
print(origin)
```

```python
# datatset의 origin 속성을 USA, Europe, Japan으로 나누어서 원핫 인코딩으로 변경
dataset['USA'] = (origin == 1)*1.0
dataset['Europe'] = (origin == 2)*1.0
dataset['Japan'] = (origin == 3)*1.0
dataset.tail()
```

```python
# 훈련 데이터와 테스트 데이터를 구분한다.
# 훈련 데이터 -> 전체 데이터의 80%
# 테스트 데이터 -> 전체 데이터의 20$
# 무작위로 선정하여 나눈다.
train_dataset = dataset.sample(frac=0.8, random_state=0)
test_dataset = dataset.drop(train_dataset.index) # 전체 데이터에서 학습데이터의 라벨은 제거한다.
```

```python
# 산점도 행렬 생성
sns.pairplot(train_dataset[["MPG", "Cylinders", "Displacement", "Weight"]], diag_kind="kde")
```

```python
# 전반적인 통계 확인
# 개수, 평균, 분산, 최소, 사분위수
# pandas.DataFrame.describe가 전반적인 통계를 보여준다.
# MPG 부분은 제거한다.
train_states = train_dataset.describe()
train_states.pop("MPG")
train_states = train_states.transpose()
train_states
```

```python
# 라벨 값은 MPG값과 같으므로 가져온다.
# MPG는 연비를 의미한다.
train_labels = train_dataset.pop('MPG')
test_labes = test_dataset.pop("MPG")
```

```python
# 데이터 정규화 함수
def norm(x):
    return (x - train_states['mean']) / train_states['std']

# 정규화 함수를 이용한 훈련, 테스트 데이터 정규화
normed_train_data = norm(train_dataset)
normed_test_data = norm(test_dataset)
```

```python
# 모델 구축
# 9 -> 64 -> 64 -> 1
# 모델 loss는 최소제곱오차를 사용, loss에는 RMSprop을 사용
def build_model():
    model = keras.Sequential([
        layers.Dense(64, activation='relu', input_shape=[len(train_dataset.keys())]),
        layers.Dense(64, activation='relu'),
        layers.Dense(1)
    ])

    optimizer = tf.keras.optimizers.RMSprop(0.001)

    model.compile(loss='mse',
                optimizer=optimizer,
                metrics=['mae', 'mse'])
        
    return model
```

```python
# 모델 생성
model = build_model()
```

```python
model.summary()
```

```python
# 훈련데이터에서 10개의 데이터를 가져온 후 predict에 사용
example_batch = normed_train_data[:10]
example_result = model.predict(example_batch)
example_result
```

```python
# 에포크가 끝날때마다 .을 찍는 콜백 클래스 생성
class PrintDot(keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs):
        if epoch % 100 == 0: print('')
        print('.', end='')
        
EPOCHS = 1000

history = model.fit(
  normed_train_data, train_labels,
  epochs=EPOCHS, validation_split = 0.2, verbose=0,
  callbacks=[PrintDot()])
```

```python
# 훈련 과정을 시각화
hist = pd.DataFrame(history.history)
hist['epoch'] = history.epoch
hist.tail()
```

```python
import matplotlib.pyplot as plt

def plot_history(history):
    hist = pd.DataFrame(history.history)
    hist['epoch'] = history.epoch
    
    plt.figure(figsize=(8,12))
    plt.subplot(2,1,1)
    plt.xlabel('Epoch')
    plt.ylabel('Mean Abs Error [MPG]')
    plt.plot(hist['epoch'], hist['mae'],
           label='Train Error')
    plt.plot(hist['epoch'], hist['val_mae'],
           label = 'Val Error')
    plt.ylim([0,5])
    plt.legend()

    plt.subplot(2,1,2)
    plt.xlabel('Epoch')
    plt.ylabel('Mean Square Error [$MPG^2$]')
    plt.plot(hist['epoch'], hist['mse'],
           label='Train Error')
    plt.plot(hist['epoch'], hist['val_mse'],
           label = 'Val Error')
    plt.ylim([0,20])
    plt.legend()
    plt.show()

plot_history(history)
# 훈련과정을 시각화한 결과 모델의 훈련이많이 진행되면 될 수록 성능이 향상되지 않는다(오버피팅 가능성)
# 따라서 가장 좋을때에서 훈련을 종료하도록 모델을 훈련을 다시 설정
```

```python
model = build_model()

# patience -> 성능 향상을 체크할 에포크 횟수
early_stop = keras.callbacks.EarlyStopping(monitor='val_loss', patience=10)

history = model.fit(normed_train_data, train_labels, epochs=EPOCHS,
                    validation_split = 0.2, verbose=0, callbacks=[early_stop, PrintDot()])

plot_history(history)
# 설정한 콜백 함수로 50번째 훈련에서 종료한다. 
```

```python
# 테스트 셋에 대한 성능 측정
loss, mae, mse = model.evaluate(normed_test_data, test_labes, verbose=2)
print("테스트 세트의 평균 절대 오차: {:5.2f} MPG".format(mae))
```

```python
# 테스트 셋에 있는 샘플로 MPG값을 예측

test_predictions = model.predict(normed_test_data).flatten()

plt.scatter(test_labes, test_predictions)
plt.xlabel('True Values [MPG]')
plt.ylabel('Predictions [MPG]')
plt.axis('equal')
plt.axis('square')
plt.xlim([0,plt.xlim()[1]])
plt.ylim([0,plt.ylim()[1]])
_ = plt.plot([-100, 100], [-100, 100])
```

```python
# 오차 분포 확인
error = test_predictions - test_labes
plt.hist(error, bins = 25)
plt.xlabel("Prediction Error [MPG]")
_ = plt.ylabel("Count")
```
